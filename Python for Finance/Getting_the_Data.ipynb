```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import datetime as dt
import yfinance as yf
```

```python
style.use('ggplot')
```

```python
start = dt.datetime(2010, 1, 1)
end = dt.datetime(2023, 11, 30)
```

```python
df_APPL = yf.download('AAPL',start, end)
df_APPL['Stock'] = ['APPL']*df_APPL.shape[0]
print(df_APPL.shape)
df_APPL.head()
```

```python
df_GS = yf.download('GS',start, end)
df_GS['Stock'] = ['GS']*df_GS.shape[0]
print(df_GS.shape)
df_GS.head()
```

```python
df = pd.concat([df_APPL, df_GS])
print(df.shape)
df.head()
```

```python
df['Date'] = df.index
df.set_index([pd.Index([x for x in range(1,7003)])], inplace=True)
print(df.shape)
df.head()
```
